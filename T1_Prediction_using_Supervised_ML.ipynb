```python
#Importing of libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading of the dataset
url = "http://bit.ly/w-data"
df = pd.read_csv(url)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum() #Finding of the missing values
```

```python
X = df.drop("Scores", axis = 1) #Dropping of Scores column in X axis
y = df["Scores"]

X #Showing of the Hours
```

```python
y #Showing of the Scores
```

```python
#Scatter Plot showing of the Relationship between Hours and Scores
sns.set_style('darkgrid')
sns.scatterplot(x= 'Hours',y='Scores',data=df)
plt.title('Relationship Between Hours and Scores')
plt.xlabel('Hours Studied', )
plt.ylabel('Scores', )
plt.show()
```

```python
#Importing of packages related to prediction in Supervised ML
#Predicted Using Linear Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
y_pred=linreg.predict(X_test)
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,linreg.predict(X),color='red')
plt.show()
```

```python
print(linreg.score(X_test,y_test))
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```

```python
pd.DataFrame({"Actual":y_test, "Predicted":y_pred})
```

What will be predicted score if the students studies for 9.25 hrs per day?

```python
hrs = pd.DataFrame([9.25])
score_prediction = linreg.predict(hrs)
score_prediction
```

According to the Linear Regression model if a student studies for 9.25 hours a day he/she is likely to score 93.89 marks.
